```python
import pandas as pd
student={'Unit Test-1':[5,6,8,3,10],'Unit Test-2':[7,8,9,6,15]}
student1={'Unit Test-1':[3,3,6,6,8],'Unit Test-2':[5,9,8,10,5]}
ds=pd.DataFrame(student)
ds1=pd.DataFrame(student1)
print(ds)
print("\n")

print(ds1)
print("\n")

print("Subtraction")
print(ds.sub(ds1))
print("\n")

print("rsub")
print(ds.rsub(ds1))
print("\n")

print("Addition")
print(ds.add(ds1))
print("\n")

print("radd")
print(ds.radd(ds1))
print("\n")

print("Multiplication")
print(ds.mul(ds1))
print("\n")

print("Division")
print(ds.div(ds1))
```

```python
import pandas as pd
grade={'Name':['Rashmi','Harsh','Ganesh','Priya','Vivek','Anita','Karthik'],'Grade':['A1','A2','B1','A1','B2','A2','A1']}
Gr=pd.DataFrame(grade,columns=['Name','Grade'])
print(Gr)
print("\n")
print(Gr.iloc[0:5])
print("\n")
print(Gr[0:5])
print("\n")
Gr['Percentage']=[92,89,None,95,68,None,93]
print(Gr)
print("\n")
Gr=Gr.reindex(['Name','Percentage','Grade'],axis=1)
print(Gr)
print("\n")
Gr=Gr.drop(['Grade'],axis=1)
print(Gr)
print("\n")
Gr=Gr.drop([2,4])
print(Gr)
```
